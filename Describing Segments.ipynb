```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Datasets/modified_clustered_data.csv')
```

```python
data.head()
```

```python
#PLOT FONT
csfont = {'fontname':'Comic Sans MS'}
```

```python
plt.figure(figsize=(25,10))

plt.title('Visits vs Gender divided by clusters',**csfont, fontsize=20)

plot = sns.barplot(x='Gender', y='Visits/year',
                    hue='cluster', data=data, 
                    ci=None)
#setting legend size
plt.setp(plot.get_legend().get_texts(), fontsize='22') 
plt.setp(plot.get_legend().get_title(), fontsize='32')
#setting labels
plt.ylabel("Visits per year", **csfont, fontsize = 15)
plt.xlabel("Gender", **csfont, fontsize = 15)
#adding counts
for i in plot.containers:
    plot.bar_label(i,)
```

```python
plt.figure(figsize=(15,10))

plt.title('Age vs like divided on clusters',**csfont, fontsize=20)

plot = sns.lineplot(x='Like', y='Age',
                     hue='cluster',
                    data=data)
#setting legend size
plt.setp(plot.get_legend().get_texts(), fontsize='22') 
plt.setp(plot.get_legend().get_title(), fontsize='32')
#setting labels
plt.ylabel("Age", **csfont, fontsize = 15)
plt.xlabel("Like", **csfont, fontsize = 15)
```

```python
plt.figure(figsize=(25,10))

plt.title('Age vs spice divided on clusters',**csfont, fontsize=20)

plot = sns.barplot(x='Age', y='spicy',
                    hue='cluster', data=data,
                     ci=None)
#setting legend
plt.setp(plot.get_legend().get_texts(), fontsize='22') 
plt.setp(plot.get_legend().get_title(), fontsize='32')
#setting labels
plt.ylabel("Spicy", **csfont, fontsize = 15)
plt.xlabel("Age", **csfont, fontsize = 15)
```
